```python
data_final = data.drop(['Country Code',	'Indicator Name',	'Indicator Code'], axis=1)
```

```python
chosen_country = 'United States'
country_data=data_final[data_final['Country Name']==chosen_country]
print(country_data)
```

```python
# filtering non numeric columns
years = country_data.columns[2:][country_data.columns[2:].str.isnumeric()].astype(int)
```

```python
#extracting population datas
population = country_data.iloc[0, 2:][country_data.columns[2:].str.isnumeric()].astype(int)
```

```python
# plotting the graph
plt.figure(figsize=(10, 6))
plt.bar(years, population, color='skyblue')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title(f'Population of {chosen_country} Over the Years')
plt.show()
```

```python
plt.hist(population, bins=25, color='skyblue', edgecolor='black')
plt.show()
```
